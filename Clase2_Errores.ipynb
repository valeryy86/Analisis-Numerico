## Medidas De Error:
 Error Absoluto: 



 Error Relativo:



 Error Porcentual:




-Ve = valor exacto. Ejemplo de notacion (P)

-Va= valor aproximado. Ejemplo de notacion (P*)


```python
Pe=0.000034
Pa=0.000054

def errores(Pe,Pa):
    Ea = (abs(Pe-Pa))
    print(f"el error absoluto es: {Ea}")
    
    Eap = (Ea * 100)
    print(f"el error absoluto porcentual es: {Eap}%")
    
    Er = (abs(Pe-Pa)/abs(Pe))
    print(f"el error relativo es: {Er}")
    
    Erp = (abs(Pe-Pa)/ abs(Pe) * 100)
    print(f"el error relativo porcentual es: {Erp}%")

errores(0.000034, 0.000054)



```
